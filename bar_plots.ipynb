```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.ticker as mtick

import glob
```

```python
UC_city_names = ['Santa Cruz', 'San Diego', 'Berkeley', 'Santa Barbara', 'Irvine', 'Davis', 'Merced', 'Riverside',
                 'Los Angeles']
```

```python
data_location = '/Users/fardila/Documents/GitHub/COLA/data/'
Zillow_data_location = data_location+'Zillow/'
```

```python
Zillow_data_files = glob.glob(Zillow_data_location+'*')
Zillow_data_files
```

```python
two_bdrm_DF = pd.read_csv(Zillow_data_files[0])
two_bdrm_DF
```

```python
UC_two_bdrm_DF = two_bdrm_DF.loc[two_bdrm_DF['RegionName'].isin(UC_city_names)]
UC_two_bdrm_DF = UC_two_bdrm_DF.sort_values('2019-10')
#add row for average
UC_two_bdrm_DF.loc['Average'] = UC_two_bdrm_DF.mean()
UC_two_bdrm_DF.loc['Average', 'RegionName'] = 'AVERAGE'
UC_two_bdrm_DF
```

```python
three_bdrm_DF = pd.read_csv(Zillow_data_files[2])
UC_three_bdrm_DF = three_bdrm_DF.loc[three_bdrm_DF['RegionName'].isin(UC_city_names)]
UC_three_bdrm_DF = UC_three_bdrm_DF.sort_values('2019-10')

#add row for average
UC_three_bdrm_DF.loc['Average'] = UC_three_bdrm_DF.mean()
UC_three_bdrm_DF.loc['Average', 'RegionName'] = 'AVERAGE'
UC_three_bdrm_DF
```

## Plot Monthly rent for all campuses 

```python
#get colors 
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

currency_format = '${x:,.0f}'
```

```python
UC_monthly_pay_no_summer = 1826
UC_monthly_pay_no_summer_array = np.array([UC_monthly_pay_no_summer for i in range(len(UC_two_bdrm_DF))])
```

```python
def make_UC_wide_bar_plot(df, n_bedrooms):
    oct2019_per_room = df['2019-10']/n_bedrooms

    #salary to prevent rent burden
    oct2019_per_room_50 = np.array(oct2019_per_room * 2)
    oct2019_per_room_33 = np.array(oct2019_per_room * 3)

    #COLA
    UC_COLA_33 = oct2019_per_room_33 - UC_monthly_pay_no_summer_array
    UC_COLA_50 = oct2019_per_room_50 - UC_monthly_pay_no_summer_array

    #make negative colas 0
    UC_COLA_33 = np.array([0 if cola <0 else cola for cola in UC_COLA_33 ])
    UC_COLA_50 = np.array([0 if cola <0 else cola for cola in UC_COLA_50 ])
    
    print(UC_COLA_33)
    ####
    #plot
    ####
    
    fig, ax = plt.subplots(1, 1, figsize=(30, 23))
    ax.set_axisbelow(True)
    ind = np.arange(len(UC_two_bdrm_DF))    # the x locations for the groups
    width = 0.35       # the width of the bars: can also be len(x) sequence

    p1 = plt.bar(ind, oct2019_per_room, width,
                 label = '1/{0} of Median Monthly rent in {0} bedroom unit (Oct 2019, Zillow)'.format(n_bedrooms),
                color = colors[0])

    plt.plot(ind,UC_monthly_pay_no_summer_array, c=colors[3], linewidth=15, label = 'Monthly TA Salary (no summer stipend) - $1,826')

    plt.scatter(ind[UC_COLA_33 ==0], oct2019_per_room_33[UC_COLA_33 ==0], marker='_', s=2500, linewidth=5,
                zorder=10, color=colors[2])
    p2 = plt.bar(ind, UC_COLA_33, width, bottom=UC_monthly_pay_no_summer_array, color=colors[2],
                 label='COLA for 33% rent burden')

    plt.scatter(ind[UC_COLA_50 ==0], oct2019_per_room_50[UC_COLA_50 ==0], marker='_', s=2500, linewidth=5,
                zorder=10, color=colors[4])
    p3 = plt.bar(ind, UC_COLA_50, width*0.65, bottom=UC_monthly_pay_no_summer_array, color=colors[4],
                 label='COLA for 50% rent burden')

    #text labels
    for i in ind:
        if UC_COLA_50[i] >0:
            plt.text(i,UC_monthly_pay_no_summer+UC_COLA_50[i]/2., '${:,.0f}'.format(int(UC_COLA_50[i])), color='w',
                     rotation=90, fontsize=25, weight='bold', horizontalalignment='center', verticalalignment='center')
        if UC_COLA_33[i] > 0:
            plt.text(i,UC_monthly_pay_no_summer+UC_COLA_33[i]*0.65, '${:,.0f}'.format(int(UC_COLA_33[i])), color='w',
                     rotation=90, fontsize=6*np.log(UC_COLA_33[i]), weight='bold', horizontalalignment='center', verticalalignment='center')
            plt.text(i,UC_monthly_pay_no_summer+UC_COLA_33[i]*0.65, '${:,.0f}'.format(int(UC_COLA_33[i])), color='k',
                     rotation=90, fontsize=6.5*np.log(UC_COLA_33[i]), weight='bold',
                     horizontalalignment='center', verticalalignment='center', zorder=0)

    ax.grid(linestyle='-', axis='y', linewidth='1', color='grey')


    tick = mtick.StrMethodFormatter(currency_format)
    ax.yaxis.set_major_formatter(tick) 
    plt.yticks(fontsize = 45)
    plt.xticks(ind, UC_two_bdrm_DF['RegionName'], fontsize = 45,rotation=90)
    plt.legend(fontsize=40, bbox_to_anchor=(0,1.02,0,1.02), loc='lower left')
    
    url = "https://github.com/f-ardila/COLA"
    txt = plt.text(-1,-1200, 'Data + code: github.com/f-ardila/COLA', url=url, fontsize=20, color='#0000EE',
                   bbox = dict(color='#0000EE', alpha=0.01, url=url))
    
    plt.savefig('figures/UC_bar_plot_COLAs_{0}bedroom.pdf'.format(n_bedrooms), bbox_inches='tight')
    plt.show()
```

```python
 make_UC_wide_bar_plot(UC_two_bdrm_DF, 2)
```

```python
 make_UC_wide_bar_plot(UC_three_bdrm_DF, 3)
```
